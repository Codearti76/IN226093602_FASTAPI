```python
import re
import string
import nltk

# Download required resources
nltk.download('stopwords')
nltk.download('wordnet')


from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english')) # unique stopwords
lemmatizer = WordNetLemmatizer()
```

```python

def preprocess_text(text):
   

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove Emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove Numbers
    text = re.sub(r"\d+", "", text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]

    # Remove short tokens (≤2 except 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    # Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return tokens
```

# **TASK 3:STRESS TESTING**



```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_sentences:
    tokens = preprocess_text(text)
    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", " ".join(tokens))
```

TASK 4 : TOKEN ANALYTICS

**2. Which sentence retained the most meaningful tokens after cleaning?**



```python
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, round(avg_token_length, 2)
for text in test_sentences:
    tokens = preprocess_text(text)
    total, unique, avg = token_analytics(tokens)

    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print(f"Total Tokens: {total}")
    print(f"Unique Tokens: {unique}")
    print(f"Average Token Length: {avg}")
```

**Task 5: Frequency Analysis Combine all tokens and use: from collections import Counter**

```python
from collections import Counter
all_tokens = []

for text in test_sentences:
    tokens = preprocess_text(text)
    all_tokens.extend(tokens)
word_counts = Counter(all_tokens)
top_10 = word_counts.most_common(10)

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word}: {count}")
least_5 = sorted(word_counts.items(), key=lambda x: x[1])[:5]

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word}: {count}")
print("\n--- Top 10 Words ---")
print(top_10)

print("\n--- Least 5 Words ---")
print(least_5)
```

## Task 6 & Task 7: Full Pipeline and Error Handling

The full pipeline processes a list of text inputs and returns:
- A combined list of all tokens
- Cleaned sentences

Error handling ensures:
- Empty strings return empty output
- Emoji-only text is cleaned safely
- Number-only text results in no tokens
- Non-string inputs are handled gracefully

This makes the pipeline robust and production-ready.

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        # Handle None or non-string input
        if not isinstance(text, str):
            tokens = []
        else:
            tokens = preprocess_text(text)

        # Store tokens
        all_tokens.extend(tokens)

        # Create cleaned sentence
        clean_sentence = " ".join(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for sent in result["clean_sentences"]:
    print(sent)
```
